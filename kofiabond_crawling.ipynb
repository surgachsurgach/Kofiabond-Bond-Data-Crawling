```python
import requests
import pandas as pd
from bs4 import BeautifulSoup

# 채권종류 선택 리스트
gen_req_url = 'http://www.kofiabond.or.kr/proframeWeb/XMLSERVICES/'
form_data_bnd = '''
                <message>
                  <proframeHeader>
                    <pfmAppName>BIS-KOFIABOND</pfmAppName>
                    <pfmSvcName>BISBndSrtPrcSrchSO</pfmSvcName>
                    <pfmFnName>selectBndTyp</pfmFnName>
                  </proframeHeader>
                  <systemHeader></systemHeader>
                    <BISBndSrtPrcTrmDTO>
                </BISBndSrtPrcTrmDTO>
                </message>
                '''

r = requests.post(gen_req_url, form_data_bnd)
soup = BeautifulSoup(r.content, 'lxml-xml')
col_dt = {
    'val1': 'bond_cd',
    'val2': 'bond_nm',
    'val3': 'bond_snm',
    'val4': 'bond_enm',
    'val5': 'gyn'
}
res_li = soup.find_all('BISBndSrtPrcTrmDTO')

df = pd.DataFrame()
for r in res_li:
    dt = dict()
    for i in list(col_dt.keys()):
        dt[i] = [r.find(i).text]
    df_sub = pd.DataFrame.from_dict(dt)
    df = df.append(df_sub, sort=True)

df.reset_index(drop=True, inplace=True)
df = df.rename(columns=col_dt)
bnd_df = df.reindex(columns=list(col_dt.values()))
```

```python
# 채권만기 선택 리스트
form_data_mat = '''
                <message>
                  <proframeHeader>
                    <pfmAppName>BIS-COM</pfmAppName>
                    <pfmSvcName>BISComCommonCodeSO</pfmSvcName>
                    <pfmFnName>selectCommonCode</pfmFnName>
                  </proframeHeader>
                  <systemHeader></systemHeader>
                    <BISComCommonCodeDTO>
                    <vGrpCdList>'C8002'</vGrpCdList>
                </BISComCommonCodeDTO>
                </message>
                '''
r = requests.post(gen_req_url, form_data_mat)
soup = BeautifulSoup(r.content, 'lxml-xml')
col_dt = {
    'groupCd': 'mat_gcd',
    'commonCd': 'mat_cd',
    'codeNm': 'mat_nm',
    'codeEngNm': 'mat_enm',
    'defaultYN': 'dyn',
}

res_li = soup.find_all('BISComCommonCodeDTO')

df = pd.DataFrame()
for r in res_li:
    dt = dict()
    for i in list(col_dt.keys()):
        dt[i] = [r.find(i).text]
    df_sub = pd.DataFrame.from_dict(dt)
    df = df.append(df_sub, sort=True)

df.reset_index(drop=True, inplace=True)
df = df.rename(columns=col_dt)
mat_df = df.reindex(columns=list(col_dt.values()))
```

```python
#
#   크롤링
#
df_all = pd.DataFrame() #데이터 저장할 데이터프레임 지정
def kofiabond_Crawling(fr_date,to_date):
    import requests
    import pandas as pd
    from bs4 import BeautifulSoup
    
    #우리가 원하는 정보가 있는 tag를 사전에 저장해둡니다.

    col_dt = { 
        'standardDt': 'date',  
        'val2': 'avg',
        'val3': 'nic',
        'val4': 'kor',
        'val5': 'kis',
        'val6': 'fnp',
    }
    
    #변경해줄 컬럼 명 리스트
    col_order = ['date', 'bnd_cd', 'mat_cd', 'avg', 'nic', 'kor', 'kis', 'fnp']

    for bnd_type in list(bnd_df['bond_cd'])[:]:        # 모든 채권 유형에 대해
        for mat_type in list(mat_df['mat_cd']):     # 모든 채권 만기에 대해

            # 조회하기
            form_data_src = '''
                            <message>
                              <proframeHeader>
                                <pfmAppName>BIS-KOFIABOND</pfmAppName>
                                <pfmSvcName>BISBndSrtPrcSrchSO</pfmSvcName>
                                <pfmFnName>selectTrm</pfmFnName>
                              </proframeHeader>
                              <systemHeader></systemHeader>
                                <BISBndSrtPrcTrmDTO>
                                <val31>bnd_cd_var</val31>
                                <val32>mat_cd_var</val32>
                                <creditEstOrgCd>comp_cd_var</creditEstOrgCd>
                                <standardDt1>fr_date_var</standardDt1> 
                                <standardDt2>to_date_var</standardDt2> 
                            </BISBndSrtPrcTrmDTO>
                            </message>
                            '''
            
            #원하는 입력값으로 대체
            form_data_src = form_data_src.replace('bnd_cd_var', bnd_type)\
            .replace('mat_cd_var', mat_type) .replace('comp_cd_var', '0')\
            .replace('fr_date_var', str(fr_date))\
            .replace('to_date_var', str(to_date)) 
            
            #request
            r = requests.post(gen_req_url, form_data_src)
            soup = BeautifulSoup(r.content, 'lxml-xml')
            res_li = soup.find_all('BISBndSrtPrcTrmDTO')   #BISBndSrtPrcTrmDTO 라는 큰 tag 안에 작은 tag들이 포함되어 있는 형태
            
            #파싱
            df = pd.DataFrame()
            for r in res_li:
                dt = dict()
                for i in list(col_dt.keys()):
                    dt[i] = [r.find(i).text]   #BISBndSrtPrcTrmDTO tag 안에 작은 tag들이 가지고 있는 정보 text로 반환
                df_sub = pd.DataFrame.from_dict(dt)
                df = df.append(df_sub, sort=True)  
                

            df['bnd_cd'] = bnd_type
            df['mat_cd'] = mat_type
            df_all.append(df, sort=True) #최종 저장 데이터프레임
            print(bnd_type, mat_type, fr_date, to_date)

    df_all.reset_index(drop=True, inplace=True)
    df_all.rename(columns=col_dt) 

    # 문자형 숫자 -> 숫자형 숫자로
    #col_numeric = ['date', 'avg', 'nic', 'kor', 'kis', 'fnp']
    #for i in col_numeric:
    #    df_all[i] = pd.to_numeric(df_all[i], errors='coerce')
        
    df_all.reindex(columns=col_order)
```

```python
#데이타 프레임의 날짜 리스트 뽑아오기
date_list = list(data.date)
date_list
#date 컬럼 삭제
del data['date']
#데이타 프레임의 인덱스를 date_list로 설정
data.index = date_list
```

```python
#반복문으로 해결하기
mat = '1 3 6 9 100 106 200 206 300 400 500 700 1000 2000 3000 5000'
mat_list = mat.split(' ')   
total_df =pd.DataFrame()

for i,mat in enumerate(mat_list):
    temp_df=data[data['mat_cd']==int(mat)]
    del temp_df['mat_cd']
    temp_df.columns = [str(mat)]
    if i != 0:
            total_df = pd.merge(total_df, temp_df, how='outer', right_index=True, left_index=True)
    else:
            total_df = temp_df
```
